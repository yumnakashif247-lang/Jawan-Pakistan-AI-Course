```python
print("\t--- Calculator ---")
a=int(input("Enter value of a:"))
b=int(input("Enter the value of b:"))
operator=input("Enter the operator of your choice(+,-,*,/,power(**)):")
result=1
if operator == "+":
    print("a + b:",a+b)
elif operator == "-":
    print("a - b:",a-b)
elif operator == "*":
    print("a * b:",a*b)
elif operator == "/":
    print("a / b:",a/b)               
elif operator == "**" or operator == "power":
    if b == 0:
        print("a power b is:",result)
    else:
      for i in range(b):
         result*=a
      print("a power b is:",result)
else:
    print("Invalid operator.")
```

```python
myList=["Yumna",20,"BS(IT)-3",3.91]
found=False
for numeric in myList:
    if isinstance(numeric,(int,float)):
        found=True
if found:
    print("Numeric value is found in list.")
else:
    print("Numeric value is not found in list.")
```

```python
myCar={"Brand":"Toyota","Speed":180,"Model":"Corolla","Price":600000,"Mileage":14}
sum=0
for each_value in myCar.values():
    if isinstance(each_value,(int,float)):
        sum+=each_value
print("Sum of all numeric values in dictionary is:",sum)
```

```python
number=[1,22,67,90,22,1,5,6,7,5,6,]
duplicate=[]
for i in range(len(number)):
    for j in range(i+1,len(number)):
        if number[i]==number[j] and number[i] not in duplicate:
            duplicate.append(number[i])
print("Duplicate values are:",duplicate)

```

```python
myCar={"Brand":"Toyota","Speed":180,"Model":"Corolla","Price":600000,"Mileage":14}
key=input("Enter the key you want to append in dictionary:")
found=False
for each_key in myCar.keys():
    if key==each_key:
        found=True
if found:
    print("Key already present in dictionary.")
else:
    print("Key not present in dictionary.")
```
