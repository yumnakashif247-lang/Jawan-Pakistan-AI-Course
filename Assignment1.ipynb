```python
DSA=float(input("Enter the marks of DSA:"))
calculus=float(input("Enter the marks of Calculus:"))
PST=float(input("Enter the marks of PST:"))
OOP=float(input("Enter the marks of OOP:"))
physics=float(input("Enter the marks of Physics:"))
total=DSA+calculus+PST+OOP+physics
percentage=total/500*100
if percentage<=100 and percentage>=85:
    grade='A'
elif percentage< 85 and percentage>=80:
    grade='A-'
elif percentage< 80 and percentage>=75:
    grade='B+'
elif percentage< 75 and percentage>=70:
    grade='B'
elif percentage< 70 and percentage>=60:
    grade='B-'
elif percentage<60 and percentage>=55:
    grade='C'
elif percentage< 55 and percentage>=50:
    grade='D'
else:
    grade='F'
print("\t----Report Card----")
print("DSA:",DSA)
print("Calculus:",calculus)
print("PST:",PST)
print("OOP:",OOP)
print("Physics:",physics)
print("Total:",total)
print("Percentage:",percentage)
print("Grade:",grade)
```

```python
print("\t--- Simple Calculator ---")
a=int(input("Enter the value of a:"))
b=int(input("Enter the value of b:"))
operator=(input("Enter the operator:"))

if operator=="+":
    print("a + b:",a+b)
elif operator=="-":
    print("a - b:",a-b)
elif operator=="*":
    print("a * b:",a*b)
elif operator=="/":
    if b==0:
        print("Division Not Possible!")
    else:
        print("a / b:",a/b)
elif operator=="%":
      if b==0:
        print("Division Not Possible!")
      else:
        print("a % b:",a%b)
else:
    print("Invalid Operator.")
```
